# **A) Dentre os seguintes países nórdicos: Suécia, Dinamarca e Noruega, verifique: No século XXI (a partir de 2001), qual foi o maior medalhista de ouro, considerando apenas as seguintes modalidades:**
 
1.   Curling
2.   Patinação no gelo (skating)
3.   Esqui (skiing)
4.   Hóquei sobre o gelo (ice hockey)

```
import pandas as pd
```

```
df = pd.read_csv('Winter_Olympics_Medals.csv')
df
```

```
df = df[['Year', 'Sport', 'NOC', 'Medal']]
df
```

```
filtered_year = df['Year'] > 2001
df = df[filtered_year]
df
```

```
nordic_countries = ['NOR', 'SWE', 'DNK']
filtered_countries = df['NOC'].isin(nordic_countries)
df = df[filtered_countries]
df
```

```
filtered_medal = df['Medal'] == 'Gold'
df = df[filtered_medal]
df
```

```
sports = ['Curling', 'Skating', 'Skiing', 'Ice Hockey']
filtered_sports = df['Sport'].isin(sports)
df = df[filtered_sports]
df
```

```
df.index = range(len(df))
df
```

```
df = df.groupby(['NOC'])
df.groups
```

```
df[['Medal']].count().rename(columns = {'Medal': 'Medalhas'})
```
