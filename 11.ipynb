# Obtenha dentre os jogos do gênero de ação (Action), tiro (Shooter) e plataforma (Platform):

- Quais são as três marcas que mais publicaram jogos dos três gêneros combinados? Indique também o total de jogos de cada marca.
- Quais são as três marcas que mais venderam os três gêneros combinados? Indique também o total de vendas de cada marca.
- Qual é a marca com mais publicações em cada um dos gêneros nos últimos dez anos no Japão? Indique também o número total de jogos dela.
- Qual foi a marca que mais vendeu em cada um desses gêneros nos últimos dez anos, no Japão? Indique também o total de vendas dela.

```
import pandas as pd
```

```
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df.dropna(0, inplace=True)
dt_frame = df
df
```

```
# Filtrando pelos três gêneros
genres = ['Action', 'Shooter', 'Platform']
filtered_genre = df['Genre'].isin(genres)
df = df[filtered_genre]
df
```

## A) Quais são as três marcas que mais publicaram jogos dos três gêneros combinados? Indique também o total de jogos de cada marca.

```
df.groupby('Publisher')['Name'].count().sort_values(ascending=False).head(3)
```

## B) Quais são as três marcas que mais venderam os três gêneros combinados? Indique também o total de vendas de cada marca.

```
df.groupby(['Publisher'])['Global_Sales'].sum().sort_values(ascending=False).head(3)
```

## C) Qual é a marca com mais publicações em cada um dos gêneros nos últimos dez anos no Japão? Indique também o número total de jogos dela.

```
# Gerando um novo Dataframe para ser usado na letra C
df_letter_c = dt_frame
df_letter_c
```

```
# Filtrando pelos últimos 10 anos
filtered_by_the_last_10_years = (df_letter_c['Year_of_Release'] > 2010)
df_letter_c = df_letter_c[filtered_by_the_last_10_years]
df_letter_c
```

```
# Filtrando por onde tem venda no Japão
filtered_by_japan = df_letter_c['JP_Sales'] != 0
df_filtered_c = df_letter_c[filtered_by_japan]
df_filtered_c
```

```
# Obtendo a marca com mais publicações no gênero action
filtered_by_action = df_filtered_c['Genre'] == 'Action'
df_filtered_c_action = df_filtered_c[filtered_by_action]
action = df_filtered_c_action.groupby('Publisher')['Genre'].count().sort_values(ascending=False).head(1)
brand_name = action.index[0]
total_brand_publications = action[0]
# Criando um Dataframe filtrado pelo nome da marca com mais publicações no gênero action
filtered_by_brand_name = df_filtered_c['Publisher'] == brand_name
df_filtered_c_by_brand_name = df_filtered_c[filtered_by_brand_name]
df_filtered_c_by_brand_name
# Obtendo o total de jogos da marca
total_brand_games = df_filtered_c_by_brand_name.groupby('Publisher')['Name'].count().sort_values(ascending=False)
total_brand_games = total_brand_games[0]
```

```
print(
    'No gênero action, a marca com mais publicações foi:\n'
    f'Marca: {brand_name} - Total de publicações = {total_brand_publications} e Total de jogos = {total_brand_games}'
    )
```

```
# Obtendo a marca com mais publicações no gênero shooter
filtered_by_shooter = df_filtered_c['Genre'] == 'Shooter'
df_filtered_c_shooter = df_filtered_c[filtered_by_shooter]
shooter = df_filtered_c_shooter.groupby('Publisher')['Genre'].count().sort_values(ascending=False).head(1)
brand_name = shooter.index[0]
total_brand_publications = shooter[0]
# Criando um Dataframe filtrado pelo nome da marca com mais publicações no gênero shooter
filtered_by_brand_name = df_filtered_c['Publisher'] == brand_name
df_filtered_c_by_brand_name = df_filtered_c[filtered_by_brand_name]
df_filtered_c_by_brand_name
# Obtendo o total de jogos da marca
total_brand_games = df_filtered_c_by_brand_name.groupby('Publisher')['Name'].count().sort_values(ascending=False)
total_brand_games = total_brand_games[0]
```

```
print(
    'No gênero shooter, a marca com mais publicações foi:\n'
    f'Marca: {brand_name} - Total de publicações = {total_brand_publications} e Total de jogos = {total_brand_games}'
    )
```

```
# Obtendo a marca com mais publicações no gênero platform
filtered_by_platform = df_filtered_c['Genre'] == 'Platform'
df_filtered_c_platform = df_filtered_c[filtered_by_platform]
platform = df_filtered_c_platform.groupby('Publisher')['Genre'].count().sort_values(ascending=False).head(1)
brand_name = platform.index[0]
total_brand_publications = platform[0]
# Criando um Dataframe filtrado pelo nome da marca com mais publicações no gênero platform
filtered_by_brand_name = df_filtered_c['Publisher'] == brand_name
df_filtered_c_by_brand_name = df_filtered_c[filtered_by_brand_name]
df_filtered_c_by_brand_name
# Obtendo o total de jogos da marca
total_brand_games = df_filtered_c_by_brand_name.groupby('Publisher')['Name'].count().sort_values(ascending=False)
total_brand_games = total_brand_games[0]
```

```
print(
    'No gênero platform, a marca com mais publicações foi:\n'
    f'Marca: {brand_name} - Total de publicações = {total_brand_publications} e Total de jogos = {total_brand_games}'
    )
```

## D) Qual foi a marca que mais vendeu em cada um desses gêneros nos últimos dez anos, no Japão? Indique também o total de vendas dela.

```
"""
Dataframe filtrado por:
- Últimos 10 anos;
- Filtro por onde teve venda no japão
"""
df_filtered_c
```

```
# Filtrando pelo gênero action
filtered_by_action = df_filtered_c['Genre'] == 'Action'
df_filtered_c_action = df_filtered_c[filtered_by_action]
df_filtered_c_action
```

```
# Obtendo a marca que mais vendeu no gênero action
df_filtered_c_action.groupby(['Publisher'])['JP_Sales'].sum().sort_values(ascending=False).head(1)
```

```
# Filtrando pelo gênero shooter
filtered_by_shooter = df_filtered_c['Genre'] == 'Shooter'
df_filtered_c_shooter = df_filtered_c[filtered_by_shooter]
df_filtered_c_shooter
```

```
# Obtendo a marca que mais vendeu no gênero shooter
df_filtered_c_shooter.groupby(['Publisher'])['JP_Sales'].sum().sort_values(ascending=False).head(1)
```

```
# Filtrando pelo gênero platform
filtered_by_platform = df_filtered_c['Genre'] == 'Platform'
df_filtered_c_platform = df_filtered_c[filtered_by_platform]
df_filtered_c_platform
```

```
# Obtendo a marca que mais vendeu no gênero platform
df_filtered_c_platform.groupby(['Publisher'])['JP_Sales'].sum().sort_values(ascending=False).head(1)
```
